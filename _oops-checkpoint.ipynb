```python
# static methods
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} {self.age}"
    
    @staticmethod
    def greet() :
        print("hello")
    
p1 = Person("rahul",35)
print(p1)
Person.greet()
```

```python
# calculate percentage of the student
class Student : 
    def __init__(self,chem,physics,maths):
        self.chem = chem
        self.physics = physics
        self.maths = maths

    def Calculate_percent(self) :
        total = self.chem + self.physics + self.maths 
        percent = (total/300)*100
        return percent
    
s = Student(45,87,98)
s.Calculate_percent()
```

```python
class Shape :
    def area1(self) :
        print("Area Calculation")

class Rectangle(Shape) :
    def __init__(self,length,breadth) :
        self.length = length
        self.breadth = breadth
        area = self.length * self.breadth
        self.area = area
    
    def Show(self) :
        return f"area of rectangle : {self.area}"

class Circle(Shape) :
    def __init__(self,radius) :
        self.radius = radius 
        area = self.radius * self.radius
        self.area = area
    
    def Show(self) :
        return f"area of Circle : {self.area}"

r = Rectangle(5,4)
r.area1()
r.Show()

c = Circle(5)
c.area1()
c.Show()
```

```python
class Person:
    def __init__(self, name):
        self.name = name

    def show_person(self):
        print("Name:", self.name)


class Employee(Person):
    def __init__(self, name, emp_id):
        super().__init__(name)
        self.emp_id = emp_id

    def show_employee(self):
        print("Employee ID:", self.emp_id)


class Developer(Employee):
    def __init__(self, name, emp_id, language):
        super().__init__(name, emp_id)
        self.language = language

    def show_developer(self):
        print("Programming Language:", self.language)


d = Developer("Ankit", 101, "Python")

d.show_person()
d.show_employee()
d.show_developer()
```

```python
class University:
    def __init__(self, university_name):
        self.university_name = university_name

    def show_university(self):
        print("University:", self.university_name)


class Department(University):
    def __init__(self, university_name, department_name):
        super().__init__(university_name)
        self.department_name = department_name

    def show_department(self):
        print("Department:", self.department_name)


class Course(Department):
    def __init__(self, university_name, department_name, course_name):
        super().__init__(university_name, department_name)
        self.course_name = course_name

    def show_course(self):
        print("Course:", self.course_name)


c = Course("Indore University", "Computer Science", "Python Programming")

c.show_university()
c.show_department()
c.show_course()
```
